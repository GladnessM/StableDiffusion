```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("jzli/majicMIX-realistic-7", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "1girl, 8k, masterpiece, ultradetailed"
h=800
w=640
steps=25
guidance=8.0
neg="easynegative, lowres, bad anatomy, bad hands, text, error, missing fingers, bad feet, cropped, poory drawn face, deformed"
image = pipe(prompt, height=h, weight=w, num_inference_steps=steps, guidance_scale=guidance, negative_prompt=neg).images[0]
image
```
